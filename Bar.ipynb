xticks and yticks

Syntax:
    xticks(ticks=None,label=None,**kwargs)

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

# Customizing x-axis and y-axis ticks
plt.xticks([0, 2, 4, 6, 8, 10], ['zero', 'two', 'four', 'six', 'eight', 'ten'], rotation=45)
plt.yticks([-1, -0.5, 0, 0.5, 1], fontsize=12)

plt.show()
```

REMOVING TICKS

```python
plt.xticks([])
plt.yticks([])
```

tick_params()

Syntax:

plt.tick_params(axis='both', which='both', direction='in', length=6, width=2, color='red')


```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

# Customizing x-axis and y-axis ticks
plt.xticks([0, 2, 4, 6, 8, 10], ['zero', 'two', 'four', 'six', 'eight', 'ten'], rotation=45)
plt.yticks([-1, -0.5, 0, 0.5, 1], fontsize=12)

# Customizing tick parameters
plt.tick_params(axis='x', direction='inout', length=8, width=2, colors='blue', labelsize=12)
plt.tick_params(axis='y', direction='out', length=6, width=1.5, colors='green', labelsize=14, labelrotation=45)

plt.show()

```

text()

Syntax:

plt.text(x, y, 'Your Text', fontsize=12, color='red')


```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

# Adding text annotation
plt.text(5, 0, 'Midpoint', fontsize=14, color='red', ha='center', va='center')
plt.text(2, 0.5, 'Start Point', fontsize=12, color='blue', bbox=dict(facecolor='yellow', alpha=0.5))
plt.text(8, -0.5, 'End Point', fontsize=12, color='green', bbox=dict(facecolor='pink', alpha=0.5))
for i in range(0, 11, 2):
    plt.text(i, np.sin(i), f'({i:.1f}, {np.sin(i):.2f})', fontsize=10, color='black')



plt.show()
```

annotate()

Syntax:

plt.annotate('Text Label', xy=(x, y), xytext=(x_text, y_text),
             arrowprops=dict(arrowstyle='->', color='red'))


```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

# Annotating a specific point
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(2, 1.2),
             arrowprops=dict(arrowstyle='->', color='red'))
plt.annotate('Start', xy=(0, 0), xytext=(1, -0.5),
             arrowprops=dict(arrowstyle='->', color='blue', linewidth=2))

plt.annotate('End', xy=(10, np.sin(10)), xytext=(8, -1),
             arrowprops=dict(arrowstyle='-|>', color='green', linewidth=2))

plt.annotate('Important Point', xy=(5, 0), xytext=(6, 0.5),
             bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.5),
             arrowprops=dict(arrowstyle='fancy', color='black'))



plt.show()
```


BAR GRAPH



Syntax:

bar(x,height,width=0.8,bottom=None,align='centre',data=None,**kwargs)

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

plt.bar(categories, values, color='skyblue', edgecolor='black')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Graph')

plt.show()
```

HORIZONTAL BAR GRAPH

```python
plt.barh(categories, values, color='green', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Graph')

plt.show()
```

STACKED BAR GRAPH


```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
group1 = [10, 20, 15, 30]
group2 = [5, 10, 8, 12]
group3 = [3, 7, 5, 10]

# Plot stacked bars
plt.bar(categories, group1, color='blue', label='Group 1')
plt.bar(categories, group2, bottom=group1, color='red', label='Group 2')
plt.bar(categories, group3, bottom=[i + j for i, j in zip(group1, group2)], color='green', label='Group 3')

# Labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.legend()

plt.show()
```

MULTIPLE BAR GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
values1 = [10, 25, 15, 30]  # Group 1
values2 = [12, 20, 18, 28]  # Group 2

x = np.arange(len(categories))  # Numeric positions for categories
width = 0.4  # Bar width

# Plot bars side by side
plt.bar(x - width/2, values1, width=width, label='Group 1', color='blue')
plt.bar(x + width/2, values2, width=width, label='Group 2', color='red')

# Customizing the x-axis
plt.xticks(x, categories)

# Labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Multiple Bar Graph')
plt.legend()

plt.show()
```

HISTOGRAM

SYNATX:

hist(x,bins=None,....,**kwargs)

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)  # 1000 random values from a normal distribution

# Create histogram
plt.hist(data, bins=20, color='blue', edgecolor='black')

# Labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)  # 1000 values from a normal distribution

# Plot histograms with different bin counts
plt.figure(figsize=(12, 5))

# 1st Histogram (5 bins)
plt.subplot(1, 2, 1)
plt.hist(data, bins=5, color='blue', edgecolor='black')
plt.title('Histogram with 5 bins')

# 2nd Histogram (30 bins)
plt.subplot(1, 2, 2)
plt.hist(data, bins=30, color='red', edgecolor='black')
plt.title('Histogram with 30 bins')

plt.show()
```

```python
import seaborn as sns

sns.histplot(data, bins=20, kde=True, color='blue')  # kde=True adds a density curve

plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Histogram with Density Curve')

plt.show()
```

STACKED DISTRIBUTION

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data1 = np.random.randn(1000)  # Normally distributed data
data2 = np.random.randn(1000) + 1  # Shifted distribution
data3 = np.random.randn(1000) - 1  # Shifted in the opposite direction

# Create histogram
plt.hist([data1, data2, data3], bins=30, stacked=True, color=['blue', 'red', 'green'], edgecolor='black', label=['Data 1', 'Data 2', 'Data 3'])

# Labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')
plt.legend()

plt.show()
```
