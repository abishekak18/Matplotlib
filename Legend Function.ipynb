LEGEND FUNCTION

```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [5, 15, 20, 25, 35]

# Plot two lines with labels
plt.plot(x, y1, label="Line 1", color="blue")
plt.plot(x, y2, label="Line 2", color="red")

# Add legend
plt.legend(loc='upper left') 
plt.legend(title="Legend Title")


# Show plot
plt.show()
```

MULTIPLE LEGEND


```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 15, 25, 35, 45]
y3 = [2, 10, 20, 30, 40]

# Create figure and axis
fig, ax = plt.subplots()

# Plot multiple lines
line1, = ax.plot(x, y1, label="Line 1", color="blue")
line2, = ax.plot(x, y2, label="Line 2", color="red")
line3, = ax.plot(x, y3, label="Line 3", linestyle="dashed", color="green")

# First legend
legend1 = ax.legend(handles=[line1, line2], loc="upper left", title="First Legend")

# Second legend (added manually)
legend2 = ax.legend(handles=[line3], loc="lower right", title="Second Legend")

# Add first legend to the axis manually
ax.add_artist(legend1)

# Show plot
plt.show()
```
