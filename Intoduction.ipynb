```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<font color="red"><b><span style="font-size:20px;">Basic Example</span></b></font>


```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.show()
```

<font color="blue"><b><span style="font-size:20px;">How to name X axis and Y axis</span></b></font>


```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel("X Axis")
plt.title("Sample")
plt.ylabel("Y Axis")
plt.show()
```

<font color="red"><b><span style="font-size:20px;">Basic Example For Plot Function</span></b></font>
Syntax: plot([x],y,[fmt],**kwargs)

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
plt.plot(x, y, color='red', linestyle='--', marker='o', markersize=8, linewidth=2)
```

<font color="red"><b><span style="font-size:20px;">Multiple Lines in One Plot</span></b></font>


```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [5, 15, 10, 20, 25]

plt.plot(x, y1, label="Line 1", color="blue", linestyle="-")
plt.plot(x, y2, label="Line 2", color="red", linestyle="--")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Multiple Line Plot")
plt.legend()  # Show legend
plt.grid(True)  # Add grid lines
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [5, 15, 10, 20, 25]

plt.plot(x, y1, label="Line 1", color="blue", linestyle="-")
plt.plot(x, y2, label="Line 2", color="red", linestyle="--")

plt.yscale('log')  # Log scale for Y-axis
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Multiple Line Plot")
plt.legend()  # Show legend
plt.grid(True)  # Add grid lines
plt.show()
```
